```python
duplicates = df[df.duplicated(subset=['speaker_id'], keep=False)]  # Keeps all duplicates, not just the first occurrence
duplicate_counts = duplicates.groupby('speaker_id').size()  # Counts the duplicates for each speaker_id
print(duplicate_counts)
```

```python
# Group by 'speaker_id' and filter to ensure no speaker_id has more than 20 entries
filtered_df = df.groupby('speaker_id').head(20)

# Save the filtered DataFrame to a new CSV file
filtered_df.to_csv('filtered_dataset.csv', index=False)
```
